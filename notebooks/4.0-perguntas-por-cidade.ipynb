# 1.0 Perguntas de negócio

```python
df.columns
```

## 1.1. Qual o nome da cidade que possui mais restaurantes registrados?

```python
cidade_mais_restaurante = df.groupby('city')['restaurant_id'].count().sort_values(ascending=False).reset_index()
var_city_mais = cidade_mais_restaurante.iloc[0,1]
cidade_mais_restaurante.loc[cidade_mais_restaurante['restaurant_id'] == var_city_mais,:]
```

## 1.2. Qual o nome da cidade que possui mais restaurantes com nota média acima de 4?

```python
cidade_nota_media = df.groupby('city')['aggregate_rating'].mean().sort_values(ascending=False).reset_index()
cidade_nota_media.loc[cidade_nota_media['aggregate_rating'] > 4 ,:]
```

## 1.3. Qual o nome da cidade que possui mais restaurantes com nota média abaixo de 2.5?

```python
cidade_nota_baixa = df.groupby('city')['aggregate_rating'].mean().sort_values(ascending=False).reset_index()
cidade_nota_baixa.loc[cidade_nota_baixa['aggregate_rating'] < 2.5 ,:]
```

## 1.4. Qual o nome da cidade que possui o maior valor médio de um prato para dois?

```python
cidade_maior_valor_prato = df.groupby('city')['custo_para_dois_corrigido'].mean().sort_values(ascending=False).reset_index()
cidade_maior_valor_prato.iloc[0,0]
```

## 1.5. Qual o nome da cidade que possui a maior quantidade de tipos de culinária distintas?

```python
cidade_culinaria_distinta = df.groupby('city')['cuisines'].nunique().reset_index()
cidade_culinaria_distinta.loc[cidade_culinaria_distinta['cuisines'].idxmax()]['city']
```

## 1.6. Qual o nome da cidade que possui a maior quantidade de restaurantes que fazem reservas?

```python
city_reserva_restaurante = df.loc[df['has_table_booking'] == 1, ['city','has_table_booking']].groupby('city').sum().sort_values(by= 'has_table_booking',ascending=False).reset_index()
city_reserva_restaurante.iloc[0,0]
```

## 1.7. Qual o nome da cidade que possui a maior quantidade de restaurantes que fazem entregas?

```python
city_entregas = df.loc[df['is_delivering_now'] == 1, ['city','is_delivering_now']].groupby('city').sum().sort_values(by='is_delivering_now', ascending = False).reset_index()
city_entregas.iloc[0,0]
```

## 1.8. Qual o nome da cidade que possui a maior quantidade de restaurantes que aceitam pedidos online?

```python
city_pedidos_online = df.loc[df['has_online_delivery'] == 1,['city','has_online_delivery']].groupby('city').sum().sort_values(by='has_online_delivery', ascending=False).reset_index()
city_pedidos_online.iloc[0,0]
```
